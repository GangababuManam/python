# Python Comments

Comments are few lines in python or any other programming languages, which we write to ignore them by compilers and interpreters

Basically we write comments to give some description of what each line of our code is doing , this makes code simple to understand and resuse when necessary.

In general, Comments are useful to remember the logic or our thought process so that next time we visit our own code after a long time , we understand easily

In Python, we use the hash (#) symbol to start writing a comment.

```python
#Print Hello, world to console
print("Hello, world")
```

# DocString in python

Docstring is short form for documentation string.

In general , we can say DocString is the description statement written mostly in the beginning of any module , function or class 

```python
def triple(num):
    """
    function to triple the number
    """
    return 3 * num

print(triple(10))
```

```python
print(triple.__doc__) #Docstring is available to us as the attribute __doc__ of the function
```

# Python Indentation

1. Most of the programming languages like C, C++, Java use curly braces { } to define a block of code but python uses indentation.

2. A block of code(body of function, loop or any conditional statement) starts with indentation and ends with the first unindented line.
Basically the amount of indentation is ignorable, but it must be consistent throughout that block.

3. In general four whitespaces are used for indentation and is preferred over tabs.




```python
for x in range(5):
    print(x)
```

Indentation can be ignored in line continuation. But it's a good idea to always indent. It makes the code more readable.

```python
if True:
    print("Machine Learning")
    name="GB"
```

```python
# We can also write above condition like below statement
if True: print("Machine Learning"); name = "GB"
```

# Multi-Line Statement

In Python, end of a statement is marked by a newline character. But we can make a statement extend over multiple lines with the line continuation character (\).

```python
add_sub = 18 + 67 + 65 + \
    90 - 67 + 6 * \
    74 + 98
```

```python
print(add_sub)
```

```python
#another way is
a = (18 + 67 + 65 +
    90 - 67 + 6 *
    74 + 98)
print(a)
```

```python
#also we can put multiple statements in a single line using ;
a = 10; b = 20; c = 30  
```
